```python
def pairwise_covariance(icd1, icd2, sex, age_l, age_h):
    global patients
    pid_x = set(patients[(patients.ICD == icd1) & (patients.SEX == sex) & 
                       (patients.AGE >= age_l) & (patients.AGE < age_h)].PID)
    pid_y = set(patients[(patients.ICD == icd2) & (patients.SEX == sex) & 
                       (patients.AGE >= age_l) & (patients.AGE < age_h)].PID)
    pid_xy = pid_x & pid_y
    
    k_x = len(pid_x)
    k_y = len(pid_y)
    k_xy = len(pid_xy)
    
    global patient_counts
    n = patient_counts[(patient_counts.SEX == sex) & (patient_counts.AGE >= age_l) & 
                           (patient_counts.AGE < age_h)]['N'].sum()
        
    return (n*k_xy - k_x*k_y) / (n*(n-1))
```
