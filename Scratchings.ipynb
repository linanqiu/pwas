```python
count = 0
with sas('original_data/opd3.sas7bdat') as f:
    for line in f:
        if count % 10000 == 0:
            print(count)
        count += 1
print(count)
```

```python
import csv

icds = set()
line_count = 0

with open('csv_data/opd3.csv', 'r') as f:
    reader = csv.DictReader(f)

    for line in reader:
        line_count += 1
        icds.add(line['ICD1'])
        icds.add(line['ICD2'])
        icds.add(line['ICD3'])
        if line_count % 1000000 == 0:
            print('line_count %d' % line_count)
            print(len(icds))
```
